```python
import unicodedata

def print_unicode_samples(text, n=5):
    for i, c in enumerate(text):
        if ord(c) > 127:
            print(f'Char: {repr(c)}, Unicode: {unicodedata.name(c, "UNKNOWN")}, Index: {i}')
            n -= 1
            if n == 0:
                break

sample_texts = df['content'].dropna().sample(10, random_state=42)
for text in sample_texts:
    print_unicode_samples(text)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Explicitly add the nltk data path
nltk.data.path.append('/home/akshathm/nltk_data')

# You can skip these now if you've already downloaded via terminal
# nltk.download('punkt', download_dir='/home/akshathm/nltk_data')
# nltk.download('stopwords', download_dir='/home/akshathm/nltk_data')

# Load stopwords and handle negations
stop_words = set(stopwords.words('english'))
negations = {"no", "not", "nor", "never"}
stop_words = stop_words - negations

# Define common contractions
contractions = {
    "can't": "cannot", "won't": "will not", "n't": " not", "'re": " are", "'s": " is", "'d": " would",
    "'ll": " will", "'t": " not", "'ve": " have", "'m": " am"
}

def expand_contractions(text):
    for contraction, full_form in contractions.items():
        text = re.sub(re.escape(contraction), full_form, text)
    return text

def clean_text_nltk(text):
    text = text.lower()
    text = expand_contractions(text)
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = text.replace('!', ' _EXCLAMATION_ ').replace('?', ' _QUESTION_ ')
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z0-9_\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in stop_words and t.strip()]
    return ' '.join(tokens)

# Assuming you have a DataFrame `df` with a 'content' column
df['clean_content'] = df['content'].astype(str).apply(clean_text_nltk)
print(df[['content', 'clean_content']].head(10))
```

```python
import spacy
import emoji
import re

# Load spaCy English model
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])

def extract_emojis(text):
    return ''.join(c for c in text if c in emoji.EMOJI_DATA)

def demojize_text(text):
    return emoji.demojize(text, delimiters=(" ", " "))

def clean_and_lemmatize(text):
    # Lowercase, remove URLs, HTML, and unwanted chars (except !, ?)
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = text.replace('!', ' _EXCLAMATION_ ').replace('?', ' _QUESTION_ ')
    text = re.sub(r'[^a-zA-Z0-9_\s]', '', text)  # Remove all except alphanum, _, and space

    # Lemmatize and remove stopwords (but keep negations)
    doc = nlp(text)
    negations = {"no", "not", "nor", "never"}
    tokens = [
        token.lemma_ for token in doc
        if (token.is_alpha or token.text in ["_EXCLAMATION_", "_QUESTION_"])
        and (token.text not in nlp.Defaults.stop_words or token.text in negations)
    ]
    return ' '.join(tokens)

# Extract emojis and demojize
df['emojis'] = df['content'].astype(str).apply(extract_emojis)
df['emoji_text'] = df['emojis'].apply(demojize_text)

# Clean and lemmatize text
df['clean_content'] = df['content'].astype(str).apply(clean_and_lemmatize)

# Show a sample
print(df[['content', 'emojis', 'emoji_text', 'clean_content']].head(10))
```
