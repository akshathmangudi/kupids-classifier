```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/tinder_reviews.csv', encoding='utf-8', engine='python', on_bad_lines='warn')

print('Shape:', df.shape)
df.head()
```

```python
df.info()
```

```python
import unicodedata

def print_unicode_samples(text, n=5):
    for i, c in enumerate(text):
        if ord(c) > 127:
            print(f'Char: {repr(c)}, Unicode: {unicodedata.name(c, "UNKNOWN")}, Index: {i}')
            n -= 1
            if n == 0:
                break

sample_texts = df['content'].dropna().sample(10, random_state=42)
for text in sample_texts:
    print_unicode_samples(text)
```

```python
# Randomize the entries
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled.to_csv('../data/tinder_reviews_randomized.csv', index=False)
df_shuffled.head()
```
